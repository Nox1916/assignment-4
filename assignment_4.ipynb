```python
# Import necessary libraries
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Define clustering models
models = {
    'KMeans': KMeans(n_clusters=3, random_state=42),
    'Agglomerative': AgglomerativeClustering(n_clusters=3),
    'DBSCAN': DBSCAN(eps=0.5, min_samples=5)
}

# Create a DataFrame to store results
results_df = pd.DataFrame(columns=['Model', 'Silhouette Score', 'Davies-Bouldin Score', 'Calinski-Harabasz Score'])

# Perform clustering and evaluate performance metrics for each model
for model_name, model in models.items():
    labels = model.fit_predict(X)
    silhouette_avg = silhouette_score(X, labels)
    davies_bouldin_avg = davies_bouldin_score(X, labels)
    calinski_harabasz_avg = calinski_harabasz_score(X, labels)
    results_df = results_df.append({'Model': model_name,
                                    'Silhouette Score': silhouette_avg,
                                    'Davies-Bouldin Score': davies_bouldin_avg,
                                    'Calinski-Harabasz Score': calinski_harabasz_avg},
                                   ignore_index=True)

# Display the results table
results_df
```
